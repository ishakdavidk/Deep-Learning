```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model, preprocessing
from sklearn.metrics import mean_squared_error, r2_score

lin_dataset = [i.strip().split() for i in open("D:\\David\\Tasks\\Deep Learning\\Homework\\HW2\\No 1\\Python\\lsd.dat").readlines()]

V1 = [i[0] for i in lin_dataset]
V2 = [i[1] for i in lin_dataset]

print('Tissue Concentration = ' + str(V1))
print('Math Score = ' + str(V2))
```

```python
regr = linear_model.LinearRegression()

V1 = np.array(V1, dtype=np.float64).reshape(-1,1)
V2 = np.array(V2, dtype=np.float64).reshape(-1,1)

regr.fit(V1, V2)
print('Coefficient of determination :', str(regr.score(V1, V2)))
```

```python
m = regr.coef_
b = regr.intercept_

print('Intercept : ', b[0])
print('Slope : ', m[0][0])

b_r = np.round(b,4)
m_r = np.round(m,4)
print()
print('Equation : \nV2 = ' + str(m_r[0][0]) + ' V1 + ' + str(b_r[0]))
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 25}

plt.rc('font', **font)

plt.figure(figsize=(20, 10), dpi=300)

plt.scatter(V1, V2,  color='black')
plt.plot(V1, V1*m+b, color='red', linewidth=3)

plt.xlabel('Tissue Concentration', fontsize=30, fontweight='bold');
plt.ylabel('Math Score', fontsize=30, fontweight='bold');

plt.grid(linestyle='--')
```
