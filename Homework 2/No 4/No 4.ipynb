```python
import matplotlib.pyplot as plt
import numpy as np
import re   

n=0
for i in open("D:\\David\\Tasks\\Deep Learning\\Homework\\HW2\\No 4\\tombstone.dat").readlines():
    if n==0:
        lin_dataset = re.split(r'\s{2,}', i)
        n=1
    else:
        lin_dataset = np.vstack([lin_dataset, re.split(r'\s{2,}', i)])
    
X = [i[1] for i in lin_dataset]
Y = [i[2] for i in lin_dataset]

X = np.array(X, dtype=np.float32)
Y = np.array(Y, dtype=np.float32)

print('SO2 Concentration = ' + str(X))
print('\nSurface Recession Rate = ' + str(Y))
```

```python
xsum, ysum = 0, 0
for i in X:
    xsum = xsum+i

for i in Y:
    ysum = ysum+i
    
print('xsum = ' + str(xsum))
print('ysum = ' + str(ysum))
```

```python
xmean = xsum/len(X)
ymean = ysum/len(Y)

print('xmean = ' + str(xmean))
print('ymean = ' + str(ymean))
```

```python
cov, var = 0, 0
for i in range(len(X)):
    cov = cov + (X[i]-xmean)*(Y[i]-ymean)
    
for i in range(len(X)):
    var = var + (X[i]-xmean)*(X[i]-xmean)
    
slope = cov/var
intr = ymean - slope*xmean

print('Slope = ' + str(slope))
print('Intercept = ' + str(intr))

slope_r = np.round(slope,4)
intr_r = np.round(intr,4)
print()
print('Equation : \nY = ' + str(slope_r) + ' X + ' + str(intr_r))
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 25}

plt.rc('font', **font)

plt.figure(figsize=(20, 10), dpi=300)

plt.scatter(X, Y,  color='black')
plt.plot(X, X*slope+intr, color='red', linewidth=3)

plt.xlabel('SO2 Concentration', fontsize=30, fontweight='bold');
plt.ylabel('Surface Recession Rate', fontsize=30, fontweight='bold');

plt.grid(linestyle='--')
```
